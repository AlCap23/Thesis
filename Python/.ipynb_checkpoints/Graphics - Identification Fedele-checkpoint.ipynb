```python
#Import the needed Packages
import pandas as pd
import numpy as np
import control as cn
import matplotlib.pyplot as plt
# Matplotlib inline magic
%matplotlib inline

# Define the TUBS Colorscale for 10 Colors
# TUBS Rot, Gelb, Orange, Hellgrün, Grün, Dunkelgrün, Hellblau, Blau, Dunkelblau, Violett
TUBScolorscale=[(190/255.0,30/255.0,60/255.0,0.9),
               (255/255.0,200/255.0,42/255.0,0.9),
               (225/255.0,109/255.0,0,0.9),
               (172/255.0,193/255.0,58/255.0,0.9),
               (109/255.0,131/255.0,0,0.9),
               (0,83/255.0,74/255.0,0.9),
               (102/255.0,180/255.0,211/255.0,0.9),
               (0,112/255.0,155/255.0,0.9),
               (0,63/255.0,87/255.0,0.9),
               (138/255.0,48/255.0,127/255.0,0.9)]

# Define the height & width of the plots
height = 400 # approx 11 cm
width = 600 # approx 15 cm
# Define the Font size
FontSize = 12
```

```python
# Define a FOTD System
def FOTD(K,T,L,u,t):
    # New Array for Output
    y = np.zeros_like(u)
    for i in range(0,len(u)):
        if t[i]>=L:
            y[i] = K*(1-np.exp(-(t[i]-L)/T))*u[i]
    return y   
```

```python
# Make an input and time
t = np.linspace(0,250,2000)
u = np.ones_like(t)
y = FOTD(10,40,50,u,t)
yinf = 10*np.ones_like(t)
x0 = np.array([0,0])
x1 = np.array([50,50])
z1 = np.array([0,10])

# Plot
fig, ax = plt.subplots(figsize=(10,4))
# Contours
ax.plot(t,yinf,color="black")
ax.plot(t,y,color="black")
ax.plot(x0,z1,color="black")
ax.plot(x1,z1,color="black")
# Fills
ax.fill_between(t[np.where(t<50)],yinf[np.where(t<50)], y[np.where(t<50)], facecolor=TUBScolorscale[2])
ax.fill_between(t[np.where(t>50)],yinf[np.where(t>50)], y[np.where(t>50)], facecolor=TUBScolorscale[6])
# Labels
plt.ylabel('Output')
plt.xlabel('Time [min]')
ax.grid(True)
plt.rcParams['svg.fonttype'] = 'none'
plt.savefig("FOTD_AREA_INTERPRETATION.svg")


```

```python
# Define the controller proportional and integral gain

P1 = 0.15/K1 + (0.35 - (L1*T1)/(L1+T1)**2)*T1/K1/L1
P2 = 0.15/K4 + (0.35 - (L4*T4)/(L4+T4)**2)*T4/K4/L4
I1 = 0.35*L1+13*L1*T1**2/(T1**2+12*L1*T1+7*L1**2)
I2 = 0.35*L4+13*L4*T4**2/(T4**2+12*L4*T4+7*L4**2)
```
