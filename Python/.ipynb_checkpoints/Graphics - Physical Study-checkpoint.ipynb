```python
# Import the needed packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from Algorithms import cm2in
from Algorithms import TUBScolorscale

# Plot width and heigth in cm
plot_width = 19.
plot_height = 7.5

# Define the export folder
export_folder = "../Latex/Graphics/Physical_"
```

```python
# Open the data file
# Column names
temp_range = range(273,314,5)
column_names = ["{:03d}".format(x) for x in temp_range]+['Qdot', 'Inertia Fan', 'Inertia Valve', 'Gascooler', 'Zeta']

# Read the Data from Fan to Temperature
Fan_Temp_K = pd.read_csv('./Data/FOTDS/Fan_Temperature_K.csv', sep='  ', header = None, names = column_names)
Fan_Temp_T = pd.read_csv('./Data/FOTDS/Fan_Temperature_T.csv', sep='  ', header = None, names = column_names)
Fan_Temp_L = pd.read_csv('./Data/FOTDS/Fan_Temperature_L.csv', sep='  ', header = None, names = column_names)

# Sort for Gain -> Temp, Lag -> Temp, Delay -> Temp, Parameter -> All other stuff
dict= {'Gain': Fan_Temp_K[column_names[:-5]],
         'Lag':Fan_Temp_T[column_names[:-5]],
         'Delay':Fan_Temp_L[column_names[:-5]],
         'Parameter': Fan_Temp_K[column_names[-5:]]}
# Concat
Fan_Temp = pd.concat(dict.values(),axis=1,keys=dict.keys())

# Read the Data from Fan to Pressure
Fan_Pressure_K = pd.read_csv('./Data/FOTDS/Fan_Pressure_K.csv', sep='  ', header = None, names = column_names)
Fan_Pressure_T = pd.read_csv('./Data/FOTDS/Fan_Pressure_T.csv', sep='  ', header = None, names = column_names)
Fan_Pressure_L = pd.read_csv('./Data/FOTDS/Fan_Pressure_L.csv', sep='  ', header = None, names = column_names)
# Give the Columns
dict = {'Gain': Fan_Pressure_K[column_names[:-5]],
        'Lag':Fan_Pressure_T[column_names[:-5]],
        'Delay':Fan_Pressure_L[column_names[:-5]],
        'Parameter': Fan_Temp_K[column_names[-5:]]}
# Concat
Fan_Pressure = pd.concat(dict.values(),axis=1,keys=dict.keys())

# Read the Data from Valve to Temperature
Val_Temp_K = pd.read_csv('./Data/FOTDS/Valve_Temperature_K.csv', sep='  ', header = None, names = column_names)
Val_Temp_T = pd.read_csv('./Data/FOTDS/Valve_Temperature_T.csv', sep='  ', header = None, names = column_names)
Val_Temp_L = pd.read_csv('./Data/FOTDS/Valve_Temperature_L.csv', sep='  ', header = None, names = column_names)
# Give the Columns
dict = {'Gain': Val_Temp_K[column_names[:-5]],
        'Lag':Val_Temp_T[column_names[:-5]],
        'Delay':Val_Temp_L[column_names[:-5]],
        'Parameter': Val_Temp_K[column_names[-5:]]}
Valve_Temp = pd.concat(dict.values(),axis=1,keys=dict.keys())

# Read the Data from Valve to Pressure
Val_Pres_K = pd.read_csv('./Data/FOTDS/Valve_Pressure_K.csv', sep='  ', header = None, names = column_names)
Val_Pres_T = pd.read_csv('./Data/FOTDS/Valve_Pressure_T.csv', sep='  ', header = None, names = column_names)
Val_Pres_L = pd.read_csv('./Data/FOTDS/Valve_Pressure_L.csv', sep='  ', header = None, names = column_names)
# Give the Columns
dict = {'Gain': Val_Pres_K[column_names[:-5]],
        'Lag':Val_Pres_T[column_names[:-5]],
        'Delay':Val_Pres_L[column_names[:-5]],
        'Parameter': Val_Pres_K[column_names[-5:]]}
Valve_Pressure = pd.concat(dict.values(),axis=1,keys=dict.keys())
```

```python
# Diagonal Dominance -> Steady State
# Make a plot
plt.rcParams['svg.fonttype'] = 'none'
plt.style.use('seaborn-whitegrid')

# Make a plot
plt.clf()
fig, ax = plt.subplots(1,2, figsize=cm2in(plot_width, plot_height))

# Data Sorting for Parameter Values
length = Fan_Temp.shape[0]
# Preallocation
for scenario in range(0,length):
    x = []
    dom1 = []
    dom2 = []
    for temps in temp_range:
        # Minor Diagonal
        main1 = Fan_Temp['Gain', str(temps)][scenario]
        main2 = Valve_Pressure['Gain', str(temps)][scenario]
        minor =(Fan_Pressure['Gain', str(temps)][scenario]*Valve_Temp['Gain', str(temps)][scenario])
        dom1.append(minor/main1/main2)
        dom2.append(minor/main2/main1)
        x.append(temps)
    ax[0].scatter(x,dom1)
    ax[1].scatter(x,dom2)

plt.grid(True)
plt.legend(loc="lower right", ncol=1,  bbox_to_anchor=(1.4,0.1))
plt.show()
```

```python
# Diagonal Dominance -> Steady State
# Make a plot
plt.rcParams['svg.fonttype'] = 'none'
plt.style.use('seaborn-whitegrid')

# Make a plot
plt.clf()
fig, ax = plt.subplots(1, figsize=cm2in(plot_width, plot_height))

# Data Sorting for Parameter Values
length = Fan_Temp.shape[0]
# Preallocation
for scenario in range(0,length):
    x = []
    dom = []
    for temps in temp_range:
        # Minor Diagonal
        main = (Fan_Temp['Lag', str(temps)][scenario]+Fan_Temp['Delay', str(temps)][scenario])*(Valve_Pressure['Lag', str(temps)][scenario]+Valve_Pressure['Delay', str(temps)][scenario])
        minor = (Fan_Pressure['Lag', str(temps)][scenario]+Fan_Pressure['Delay', str(temps)][scenario])*(Valve_Temp['Lag', str(temps)][scenario]+Valve_Temp['Delay', str(temps)][scenario])
        dom.append(np.abs(main/minor))
        x.append(temps)
    ax.scatter(x,dom)

plt.grid(True)
plt.legend(loc="lower right", ncol=1,  bbox_to_anchor=(1.4,0.1))
plt.show()
```

```python
# Diagonal Dominance -> Steady State
# Make a plot
plt.rcParams['svg.fonttype'] = 'none'
plt.style.use('seaborn-whitegrid')

# Make a plot
plt.clf()
fig, ax = plt.subplots(1, figsize=cm2in(plot_width, plot_height))

# Data Sorting for Parameter Values
length = Fan_Temp.shape[0]
# Preallocation
for scenario in range(0,length):
    x = []
    L11,L12,L21,L22 = [],[],[],[]
    for temps in temp_range:
        # Minor Diagonal
        L11.append(Fan_Temp['Lag', str(temps)][scenario])
        L22.append(Valve_Pressure['Lag', str(temps)][scenario])
        L21.append(Valve_Temp['Lag', str(temps)][scenario])
        L12.append(Fan_Pressure['Lag', str(temps)][scenario])
        x.append(temps)
    ax.scatter(x,L11, color = TUBScolorscale[0])
    ax.scatter(x,L21, color = TUBScolorscale[2])
    ax.scatter(x,L12, color = TUBScolorscale[9])
    ax.scatter(x,L22, color = TUBScolorscale[6])

plt.grid(True)
plt.legend(loc="lower right", ncol=1,  bbox_to_anchor=(1.4,0.1))
plt.show()
```
